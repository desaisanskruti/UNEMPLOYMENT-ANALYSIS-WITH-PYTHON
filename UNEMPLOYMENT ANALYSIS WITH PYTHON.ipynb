```python
df.columns=['State','Date','frequency', 'Estimated Unemployment Rate' ,'Estimated Employed','Estimated Labour Participation Rate ','Region', 'longitude','latitude']
```

```python
df.head()
```

```python
selected_columns = ['Estimated Unemployment Rate' ,'Estimated Employed','Estimated Labour Participation Rate ', 'longitude','latitude']
subset_df = df[selected_columns]
subset_df.corr()
```

```python
corr = subset_df.corr()
fig,ax = plt.subplots(figsize=(4,3))
sns.heatmap(corr,annot=True,ax=ax,cmap='coolwarm')
```

```python
plt.figure(figsize=(5,5))
plt.title("Unemployment in India")

sns.histplot(
    data=df,
    x="Estimated Employed",
    hue="Region"
)
plt.show()

```

```python
plt.title("Unemployment in India")
sns.histplot(x='Estimated Unemployment Rate',hue='Region',data=df)
plt.show()             
```

```python
unemployment = df[['State','Region','Estimated Unemployment Rate']]
figure = px.sunburst(unemployment,path=['Region','State'],
                     values = 'Estimated Unemployment Rate',
                     width=700,height=600,color_continuous_scale='RdY1Gn',
                     title="Unemployment Rate in India")
figure.show()
```
